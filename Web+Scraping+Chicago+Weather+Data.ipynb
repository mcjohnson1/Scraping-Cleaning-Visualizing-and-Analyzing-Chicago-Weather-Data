```python
# For web scraping
from bs4 import BeautifulSoup
import urllib
import urllib.request

# For file creation
import os
```

```python
file = open(os.path.expanduser(r"~/Desktop/Chicago Weather Data.csv"), "wb")
file.write(b"Date,Mean Temperature,Max Temperature,Min Temperature" + b"\n")
```

```python
# Function to "make soup"
def make_soup(url):
    webpage = urllib.request.urlopen(url)
    soupdata = BeautifulSoup(webpage, "html.parser")
    return soupdata
```

```python
for year in range(1996, 2018):        # From year 1996 to 2017
    for month in range(1,13):         # From 1 - 12 : Month Jan - Dec
        date = str(year) + "/" + str(month) 
        soup = make_soup("https://www.wunderground.com/history/airport/KORD/" + date + "/1/MonthlyHistory.html?req_city=Chicago&req_state=IL&req_statename=Illinois&reqdb.zip=60290&reqdb.magic=1&reqdb.wmo=99999")
        
        Max = soup.findAll('tr')[2].findAll('td')[1].find(attrs={"class":"wx-value"}).text
        Mean = soup.findAll('tr')[3].findAll('td')[2].find(attrs={"class":"wx-value"}).text
        Min = soup.findAll('tr')[4].findAll('td')[3].find(attrs={"class":"wx-value"}).text
        
        tempString = date + "/1" + "," + Mean + "," + Max + "," + Min + "\n"
        file.write(bytes(tempString, encoding="ascii", errors='ignore'))
        
        print(tempString)

file.close()
  
```
